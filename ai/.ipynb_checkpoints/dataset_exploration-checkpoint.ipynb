```python
def normalize_gpa(gpa):
    """
    Normalizo vlerat e notes mesatare nga shkalla US 2-5
    ne shallen evropiane 6-10.
    """
    return float(((gpa - 2.0) / 3.0) * 4.0 + 6.0)
```

```python
clean_df = df.copy()

clean_df.columns = clean_df.columns.str.lower()

clean_df['gpa'] = [round(normalize_gpa(gpa), 4) for gpa in clean_df['gpa']]
```
